# Making changes to the data

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(5)
```

```python
#Drop a column

#df = df.drop(columns=['Total'])

df['Total'] = df.iloc[:, 4:10].sum(axis=1)

df.head(5)
```

```python
cols = list(df.columns.values)
df = df[cols[0:9] + [cols[-1]]+cols[10:12]]
df
```

# Conditional Changes

```python
# must specify the column anme to be changed. otherwise all the row will get changed.
#df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer' 
df.loc[df['Type 1'] == 'Flamer', 'Type 1'] = 'Fire' 
df
```

```python
#by specifying other columns also we can change the value.
df.loc[df['Type 1'] == 'Fire', 'Legendary'] = 'True' 
df.loc[df['Type 1'] == 'Fire']
```

# Aggregate Statistics (Group by)

```python
# group by the average of all the stats by Pokemon Type
df.groupby(['Type 1']).mean()
```

```python
#sort the group by result based on the column
df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
```

```python
#use multple column in group by function
df['Count'] = 1
df.groupby(['Type 1']).count()['Count']>50   # Need to check this result
#df.groupby(['Type 1', 'Type 2']).count()['Count']
```

# Working with large data Set

```python
for df in pd.read_csv('modified_data.csv', chunksize=5):
    print('CHUNK DF')
    print(df)
```

```python
new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('modified_data.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results])

new_df
```
